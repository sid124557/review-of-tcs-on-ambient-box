```python
def extract_data_from_pages(soup,page_index): 
    col=['date','tital','user','dislike','like','rating','page_index']
    data=[]
    like_data=''
    dislike_data=''
    for div_card in soup.find_all(name='div', attrs={'class':'ab_card ab_comp_review_card'}):
        try:
            date=div_card.find_all(name='span', attrs={'class':'status caption-subdued'})
            date_data=date[0].get_text()
            #print(date_data)
            #date_function()
            tital=div_card.find_all(name='h2', attrs={'class':'bold-title-s review-title'})
            #print(tital[0].get_text())
            tital_data=tital[0].get_text()
            #print(tital_data)
            user=div_card.find_all(name='div', attrs={'class':'user-wrap'})
            #print(user[0].get_text())
            user_data=user[0].get_text()
            #print(user_data)
            riv=div_card.find_all(name='span', attrs={'class':'avg-rating bold-Labels'})
            print(riv[0].get_text())
            riv_data=riv[0].get_text()
            try:
                for text_date in div_card.find_all(name='div', attrs={'class':'content'}):
                    for like in text_date.find_all(name='h3', attrs={'class':'input-fields sub-heading'}):
                        if (like.get_text()=='Dislikes'):
                            dislike_data=text_date.get_text()
                            print(dislike_data)
                        elif (like.get_text()=='Likes'):
                            like_data=text_date.get_text()
                            print(like_data)
                            #like_text_date.append(text_date.get_text())
            except Exception as e_like:
                print("exception_in_like=============>",e_like)
            data.append([date_data,tital_data,user_data,dislike_data,like_data,riv_data,page_index])
            print('____________like______dislike_________________')        
            print(like_data,dislike_data)
            print('^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^') 
        except Exception as e:
            print("exception=============>",e)
    df=pd.DataFrame(data,columns=col)
    return(df)
```

```python
start=[2,3,4,5,6,7,8,9,10]
```

```python
import random 
dfx=[]
#s = requests.session()
#s.headers['User-Agent'] = 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_9_2) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/34.0.1847.131 Safari/537.36'
#res = s.get(my_URL)
#soup = bs4.BeautifulSoup(res.text, 'html.parser')
for s1 in start:
    s = requests.session()
    s.headers['User-Agent'] = 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_9_2) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/34.0.1847.131 Safari/537.36'
    url3='https://www.ambitionbox.com/reviews/tcs-reviews?page='+str(s1)
    #requests.headers['User-Agent'] = 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_9_2) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/34.0.1847.131 Safari/537.36'
    #page=requests.get(url3)
    res = s.get(url3)
    soup = BeautifulSoup(res.text, 'html.parser')
    #print(soup.prettify()
    if int(s1)%2==0:
        t=r1 = random.randint(0, 30) 
        print("time sleep=",t)
        time.sleep(t)
    dfx.append(extract_data_from_pages(soup,s1))
```

```python
df=pd.concat(dfx,ignore_index=True,sort=False)
```
